```python
import os
import re
import time
from typing import List, Any, Union
import pandas as pd
import multiprocessing as mp
import copy as cp
```

```python
file_dir = '../input/project/'
file_name = "train_data.csv"

dictNodes = {}

def intersection(lst1, lst2): 
    lst3 = [value for value in lst1 if value in lst2] 
    return len(lst3)

def Union(lst1, lst2): 
    final_list = list(set(lst1) | set(lst2)) 
    return final_list 

#get out
def getFollowees(node, dataframe): 
    followees = []
    count=0
    for index, row in dataframe.iterrows():
        if (node == row['UserA']):  
            count+=1
            followees.append(row['UserB'])
            
    return count, followees
        
#get in        
def getFollowers(node, dataframe):
    followers = []
    count=0
    for index, row in dataframe.iterrows():
        if (node == row['UserB']): 
            count+=1
            followers.append(row['UserA'])
            
    return count, followers






with open(os.path.join(file_dir, file_name)) as f:
    df = pd.read_csv(f,header = 0)
    
    for i in range(len(df)):
        node_out_A,node_followees_A = getFollowees(df.loc[i,'UserA'], df)
        node_in_A ,node_followers_A = getFollowers(df.loc[i,'UserA'], df)
        neighobrs_A = Union(node_followees_A,node_followers_A)
        node_out_B,node_followees_B = getFollowees(df.loc[i,'UserB'], df)
        node_in_B ,node_followers_B = getFollowers(df.loc[i,'UserB'], df)
        neighobrs_B = Union(node_followees_B,node_followers_B)
        commonNeighbors = intersection(neighobrs_A, neighobrs_B)
        df.loc[i,'UserAIn'] = node_in_A
        df.loc[i,'UserAOut'] = node_out_A
        df.loc[i,'UserBIn'] = node_in_B
        df.loc[i,'UserBOut'] = node_out_B
        df.loc[i,'ComNeighbors'] = commonNeighbors
        
    df.to_csv("sample.csv", index=False)    
    
```
